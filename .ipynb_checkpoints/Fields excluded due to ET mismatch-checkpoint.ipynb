This notebook is used to assess fields with high disaggrement between reported Irrigation 
and remotely sensed ET.

 `Background`
- We have a total of 92 yield observations and 103 reported irrigation observations from 2008 to 2022 across all fields. Not all fields were used to build simulation or predictive models, some were screened out to ensure only high-quality data were included. Following Ott et al. (2024) and Zipper et al. (2024b), we excluded 25 irrigation observations in which the ratio of irrigation to precipitation deficit was less than 0.5 or greater than 1.6, using the following equations:
 * ratio =irrigation / precipitation deficit					                        
 * precipitation deficit = evapotranspiration (ET) – effective precipitation 	      
 * effective precipitation = precipitation – (runoff + deep percolation) 	          


```python


def plot_yield_comparison(
    df, 
    x_col: str = 'Reported_Yield', 
    y_col: str = 'Simulated_Yield', 
    df_type: str = "Test", 
    hue: str = None,
    style: str = None,
    metric: bool = None,
    x_axis: str = None,
    y_axis: str = None,
    xlim: tuple = (-200, 1100), 
    ylim: tuple = (-200, 1100),
    bound: tuple = None  # New parameter for bounding region
) -> tuple:
    
    """
    Plot a comparison of reported vs. simulated/predicted  with statistical evaluation.

    This function generates a scatter plot of reported vs. simulated yield values,
    overlays a 1:1 reference line, and fits a regression line to the data. It can
    optionally display performance metrics (MAE, RMSE, R², r) and highlight a
    bounded region representing acceptable model performance.

    Parameters
    ----------
    df : pandas.DataFrame
        Input dataset containing reported and simulated yield values.
    x_col : str, optional (default='Reported_Yield')
        Column name for the x-axis (reported yields).
    y_col : str, optional (default='Simulated_Yield')
        Column name for the y-axis (simulated yields).
    df_type : str, optional (default="Test")
        Label for the dataset type (e.g., "Training", "Validation", "Test").
    hue : str, optional
        Column name to group data points by color (categorical variable).
    style : str, optional
        Column name to group data points by marker style.
    metric : bool, optional (default=None)
        If True, display MAE, RMSE, R², and correlation coefficient on the plot.
    x_axis : str, optional
        Custom label for the x-axis (overrides column name).
    y_axis : str, optional
        Custom label for the y-axis (overrides column name).
    xlim : tuple, optional (default=(-200, 1100))
        Limits for the x-axis.
    ylim : tuple, optional (default=(-200, 1100))
        Limits for the y-axis.
    bound : tuple, optional
        Acceptable range for simulated yields; shaded region is drawn if specified.

    Returns
    -------
    tuple
        (fig, axes) where:
        - fig : matplotlib.figure.Figure
            The overall figure object.
        - axes : matplotlib.axes._axes.Axes
            The axes object containing the plot.
    """
    
    
    plt.rcParams["figure.dpi"] = 300
    fig, axes = plt.subplots(figsize=(8, 5))

    # 1:1 line
    axes.axline((0, 0), slope=1, label='1:1 line', color='gray', linestyle='--')
    
    # Scatter plot
    sns.scatterplot(data=df, x=x_col, y=y_col, hue=hue, ax=axes, palette='tab10', style=style, s=50, edgecolor='k')
    
    # Regression line
    sns.regplot(data=df, x=x_col, y=y_col, scatter=False, ci=None, ax=axes, color='red', label='Regression line', line_kws={'label': None})

    # Add shaded region if `bound` is specified
    if bound:
        # Create a continuous range of x values spanning the x-axis limits
        x_range = np.linspace(df[x_col].min(), df[x_col].max(), len(df))
        
        # Fill the area between the bounds for the entire x-range
        axes.fill_between(x_range, bound[0], bound[1], color='gray', alpha=0.3, label=f'Y range: {bound[0]} to {bound[1]}')

    # Set axis labels if provided
    if x_axis is not None:
        axes.set_xlabel(x_axis, fontsize=12)
    else:
        axes.set_xlabel(x_col, fontsize=12)  # Default to x_col if x_axis is not provided

    if y_axis is not None:
        axes.set_ylabel(y_axis, fontsize=12)
    else:
        axes.set_ylabel(y_col, fontsize=12)  # Default to y_col if y_axis is not provided
        
    # Set axis limits if provided
    if xlim:
        axes.set_xlim(xlim)
    if ylim:
        axes.set_ylim(ylim)

    # Compute statistics
    corr, _ = pearsonr(df[x_col].values, df[y_col].values)
    mae = mean_absolute_error(df[x_col], df[y_col])
    rmse = np.sqrt(mean_squared_error(df[x_col], df[y_col]))

    # Display metrics if required
    if metric:
        axes.annotate(f"MAE: {mae:.2f}", xy=(0.05, 0.95), xycoords='axes fraction', fontsize=11)
        axes.annotate(f"RMSE: {rmse:.2f}", xy=(0.05, 0.90), xycoords='axes fraction', fontsize=11)
        axes.annotate(f"$R^2$: {corr**2:.2f}", xy=(0.05, 0.85), xycoords='axes fraction', fontsize=11)
        axes.annotate(f"$r$: {corr:.2f}", xy=(0.05, 0.80), xycoords='axes fraction', fontsize=11)

    # Legend & title
    axes.legend(bbox_to_anchor=(1.02, 1), loc='upper left')
    axes.set_title(f"{df_type}")

    return fig, axes
```

```python
df = pd.merge(ydf,et_df, on= ["Year", "FieldID"], how = "inner" )
df = pd.merge(df,Precip_sum, on= ["Year", "FieldID"], how = "inner" ) 
df = pd.merge(df,ret_sum, on= ["Year", "FieldID"], how = "inner" ) 
df = df.dropna(subset = ["Reported_Irrigation"])
df = pd.merge(df, water_loss, on=["Year", "FieldID"], how="inner")
df["Effective Precip"] = df["Precip_sum"] - df["total_loss"]
df['precip_deficit']  = df["ET_sum"] - df["Effective Precip"] 
df['ratio']  = df["Reported_Irrigation"]/ df['precip_deficit'] 
df
```
